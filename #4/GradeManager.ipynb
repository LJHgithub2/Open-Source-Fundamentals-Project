- 조건: 5명의 학생의 세개의 교과목 (영어, C-언어, 파이썬)에 대하여 키보드로부터 학번, 이름, 영어점수, C-언어 점수, 파이썬 점수를 입력받아 총점, 평균, 학점, 등수를 계산하는 프로그램 작성
  - 객체지향 프로그램으로 수정하기
  - 입력 함수, 총점/평균 계산 함수,  학점계산 함수, 등수계산 함수, 출력 함수로 나누어 구현
  - 삽입 함수, 삭제 함수, 탐색함수(학번,이름), 정렬(총점)함수, 80점 이상 학생 수 카운트 함수
  - list와 dictionary를 사용하여 작성
    - 입력 예시
      - 이종현 2017038088
          - 80 90 100
      - 호수준 2023038045
          - 69 40 20
      - 김정연 2022038039
          - 79 55 90
      - 홍성준 2020038200
          - 90 100 100
      - 황다인 2019038088
          - 70 50 30

```python
class Student():
  def __init__(self):
    self.name = input("학생 이름을 입력하세요: ")
    self.student_num = input("학번을 입력하세요: ")
    self.score = list(map(int, input(f"{self.name} 학생의 영어, C-언어, 파이썬 점수를 입력하시오 (예: 80 90 70): ").split()))
    self.summary = {'grade': '', 'avg': -1, 'sum': -1, 'rank':-1}


  def sum_avg(self):
    for score in self.score:
      total_score = sum(self.score)
      average_score = int(total_score / len(self.score))
      self.summary['sum'] = total_score
      self.summary['avg'] = average_score

  def grade(self):
    if self.summary['avg'] >= 90:
        self.summary['grade'] = "A"
    elif self.summary['avg'] >= 80:
        self.summary['grade'] = "B"
    elif self.summary['avg'] >= 70:
        self.summary['grade'] = "C"
    elif self.summary['avg'] >=60:
        self.summary['grade'] = "D"
    else:
        self.summary['grade'] = "F"

  def rank(self, student_list):
      rank = 1
      for student in student_list:
          if self.summary['avg'] < student.summary['avg']:
              rank += 1
      self.summary['rank'] = rank
```

```python
class GradeManager():
  def __init__(self,num):
    self.student_list=list()
    for _ in range(num):
      self.student_list.append(Student())

  def calculate_sum_avg(self):
    for student in self.student_list:
      student.sum_avg()

  def calculate_grade(self):
    for i in self.student_list:
      if i.summary['avg'] == -1:
        print("평균을 먼저 구하세요")
        return False

    for student in self.student_list:
      student.grade()

  def calculate_rank(self):
    for student in self.student_list:
        if student.summary['avg'] == -1:
            print("평균을 먼저 구하세요")
            return
    for student in self.student_list:
        student.rank(self.student_list)

  def insert_student(self):
    self.student_list.append(Student())

  def delete_student(self,name):
    new_student_list = []
    for student in self.student_list:
        if student.name != name:
            new_student_list.append(student)
    self.student_list = new_student_list

  def sort(self):
    for i in self.student_list:
      if i.summary['avg'] == -1:
        print("평균을 먼저 구하세요")
        return

    self.student_list.sort(key=lambda x: x.summary['avg'], reverse=True)

  def count_students_above_80(self):
    for i in self.student_list:
      if i.summary['avg'] == -1:
        print("평균을 먼저 구하세요")
        return

    count = 0
    for i in self.student_list:
        if i.summary['avg'] >= 80:
            count +=1
            print(f"{i.name} 학생 평균 점수 : {i.summary['avg']}")
    print(f"80점이 넘는 학생의 수는{count}명이다")

  def print(self,index=-1):
    if index == -1: # 전체 출력
      for i in self.student_list:
        if i.summary['rank'] == -1:
          rank= "미정"
        else:
          rank = i.summary['rank']
        if i.summary['avg'] == -1:
          avg= "미정"
        else:
          avg = i.summary['avg']
        if i.summary['grade'] == '':
          grade= "미정"
        else:
          grade = i.summary['grade']
        print(f"{i.name}님의 각 과목별 성적은 {i.score[0]}, {i.score[1]}, {i.score[2]}이며")
        print(f"등수는 {rank}, 평균은 {avg}, 등급은 {grade}입니다")

    else :
      i = self.student_list[index]
      if i.summary['rank'] == -1:
        rank= "미정"
      else:
        rank = i.summary['rank']
      if i.summary['avg'] == -1:
        avg= "미정"
      else:
        avg = i.summary['avg']
      if i.summary['grade'] == '':
        grade= "미정"
      else:
        grade = i.summary['grade']
      print(f"{i.name}님의 각 과목별 성적은 {i.score[0]}, {i.score[1]}, {i.score[2]}이며")
      print(f"등수는 {rank}, 평균은 {avg}, 등급은 {grade}입니다")
```

```python
students = GradeManager(3)
```

```python
students.calculate_sum_avg()
students.calculate_grade()
students.calculate_rank()
```
