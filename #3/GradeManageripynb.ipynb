# 오픈소스 기초프로젝트 과제#3
- 충북대학교 소프트웨어학과 2017038088 이종현
- 2024-04-03
- 조건: 5명의 학생의 세개의 교과목 (영어, C-언어, 파이썬)에 대하여 키보드로부터 학번, 이름, 영어점수, C-언어 점수, 파이썬 점수를 입력받아 총점, 평균, 학점, 등수를 계산하는 프로그램 작성
  - 입력 함수, 총점/평균 계산 함수,  학점계산 함수, 등수계산 함수, 출력 함수로 나누어 구현
  - 삽입 함수, 삭제 함수, 탐색함수(학번,이름), 정렬(총점)함수, 80점 이상 학생 수 카운트 함수
  - list와 dictionary를 사용하여 작성
    - 입력 예시
      - 이종현 2017038088
          - 80 90 100
      - 호수준 2023038045
          - 69 40 20
      - 김정연 2022038039
          - 79 55 90
      - 홍성준 2020038200
          - 90 100 100
      - 황다인 2019038088
          - 70 50 30

```python
def input_student_info(num):
  student_info = list()
  for i in range(num):
    insert_student(student_info)
  return student_info

def calculate_sum_average(info):
  for student in info:
    total_score = sum(student[2])
    average_score = int(total_score / len(student[2]))
    student[3]['sum'] = total_score
    student[3]['avg'] = average_score


def calculate_grade(info):
  for student in info:
    if student[3]['avg'] == -1 :
      print("평균을 먼저 구하세요")
      return
    if student[3]['avg'] >= 90:
        student[3]['grade'] = "A"
    elif student[3]['avg'] >= 80:
        student[3]['grade'] = "B"
    elif student[3]['avg'] >= 70:
        student[3]['grade'] = "C"
    elif student[3]['avg'] >=60:
        student[3]['grade'] = "D"
    else:
        student[3]['grade'] = "F"


def calculate_rank(info):
  if info[0][3]['avg'] == -1 :
    print("평균을 먼저 구하세요")
    return

  info.sort(key=lambda x: x[3]['avg'], reverse=True)
  for index, student in enumerate(info):
    student[3]['rank'] = index

def print_results(info):
  check = info[0][3]
  if check['sum'] == -1 or check['avg'] == -1 or check['rank'] == -1 or check['grade'] == '':
    print("결과를 출력 할 수 없습니다")
    return

  for index, item in enumerate(info):
    print('')
    print(f"{item[0]} 학생 총점={item[3]['sum']}, 평균={item[3]['avg']:.2f}, 학점={item[3]['grade']}, 등수={item[3]['rank'] + 1}")
```

```python
def insert_student(info):
  student_name = input("이름과 학번을 입력하세요 (예: 이종현 2017038088): ").split()
  name = student_name[0]
  scores = list(map(int, input(f"{name} 학생의 영어, C-언어, 파이썬 점수를 입력하시오 (예: 80 90 70): ").split()))
  print(f"{name} 학생 영어 점수:{scores[0]}, c-언어 점수:{scores[1]}, 파이썬 점수:{scores[2]}")
  info.append([name, student_name[1], scores, {'grade': '', 'avg': -1, 'sum': -1, 'rank':-1}])

def delete_student(info,student_num):
  for sublist in info:
    if sublist[1] == student_num:
        info.remove(sublist)
        print(f"{sublist[0]} 학생 정보가 삭제되었습니다.")
        return
  print("해당 학번을 가진 학생을 찾을 수 없습니다.")

def search_student_by_id(info, student_num):
    for sublist in info:
        if sublist[1] == student_num:
            print(f"{sublist[0]} 학생 영어 점수:{sublist[2][0]}, c-언어 점수:{sublist[2][1]}, 파이썬 점수:{sublist[2][2]}")
            break
    else:
        print("해당 학번을 가진 학생을 찾을 수 없습니다.")

def sort_students_by_score(info):
  for i in info:
    if i[3]['avg'] == -1 :
      print("평균을 먼저 구하세요")
      return

  info.sort(key=lambda x: x[3]['avg'], reverse=True)

def count_students_above_80(info):
  for sublist in info:
    if sublist[3]['avg'] == -1 :
      print("평균을 먼저 구하세요")
      return
  for sublist in info:
      if sublist[3]['avg'] >= 80:
          print(f"{sublist[0]} 학생 평균 점수 : {sublist[3]['avg']}")
```

```python
student = input_student_info(3)
calculate_sum_average(student)
calculate_grade(student)
calculate_rank(student)
print_results(student)
```
