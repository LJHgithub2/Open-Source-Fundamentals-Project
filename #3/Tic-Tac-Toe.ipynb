```python
def findOptimalLocation(board):

   # 승리 가능한 위치 확인
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'  # 일단 놓아보고
                if checkWin(board, 'O'):  # 승리 가능한 경우
                    board[i][j] = ' '  # 되돌리고
                    return i, j  # 그 위치 반환
                board[i][j] = ' '  # 되돌리기
    # 사용자의 승리 방지
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'  # 일단 놓아보고
                if checkWin(board, 'X'):  # 사용자의 승리 가능한 경우
                    board[i][j] = ' '  # 되돌리고
                    return i, j  # 그 위치 방지를 위해 선택
                board[i][j] = ' '  # 되돌리기
    # 가운데 우선
    if board[1][1] == ' ':
        return 1, 1
    # 모두 없을때 남는 곳 선택
    for i in range(3):
      for j in range(3):
        if board[i][j] == ' ':
          return i,j

def checkWin(board, player):
    # 가로 체크
    for row in board:
        if row.count(player) == 3:
            return True

    # 세로 체크
    for col in range(3):
        if [board[row][col] for row in range(3)].count(player) == 3:
            return True

    # 대각선 체크
    if board[0][0] == player and board[1][1] == player and board[2][2] == player:
        return True
    if board[0][2] == player and board[1][1] == player and board[2][0] == player:
        return True

    return False

def isBoardFull(board):
    for row in board:
        if ' ' in row:  # 빈 칸이 존재하는 경우
            return False
    return True  # 빈 칸이 없는 경우 (모든 칸이 채워진 경우)

# 사용 예시
```

```python
board= [[' ' for x in range (3)] for y in range(3)]
while True:
 # 게임 보드를 그린다.
 for r in range(3):
  print(" " + board[r][0] + "| " + board[r][1] + "| "+ board[r][2])
  if (r != 2):
    print("--|--|--")
 # 사용자로부터 좌표를 입력받는다.
 x = int(input("다음 수의 x좌표를 입력하시오: "))
 y = int(input("다음 수의2 y좌표를 입력하시오: "))
 # 사용자가 입력한 좌표를 검사한다.
 if x < 0 or x > 2 or y < 0 or y > 2 or board[x][y] != ' ':
  print("잘못된 위치입니다. ")
  continue
 else:
  board[x][y] = 'X'

  if isBoardFull(board):
      print("무승부")
      break
  x, y =findOptimalLocation(board)
  board[x][y] = 'O'
  if checkWin(board, 'X'):
    print("플레이어 승")
    break
  if checkWin(board, 'O'):
    print("컴퓨터 승")
    break
```
