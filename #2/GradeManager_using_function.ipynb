# 오픈소스 기초프로젝트 과제#2
- 충북대학교 소프트웨어학과 2017038088 이종현
- 2024-03-27
- 조건: 5명의 학생의 세개의 교과목 (영어, C-언어, 파이썬)에 대하여 키보드로부터 학번, 이름, 영어점수, C-언어 점수, 파이썬 점수를 입력받아 총점, 평균, 학점, 등수를 계산하는 프로그램 작성
  - 입력 함수, 총점/평균 계산 함수,  학점계산 함수, 등수계산 함수, 출력 함수로 나누어 구현
    - 입력 예시
      - 이종현 2017038088
          - 80 90 100
      - 호수준 2023038045
          - 69 40 20
      - 김정연 2022038039
          - 79 55 90
      - 홍성준 2020038200
          - 90 100 100
      - 황다인 2019038088
          - 70 50 30

```python
def input_student_info(num):
  student_info = list()
  for i in range(num):
    student_name = input("이름과 학번을 입력하세요 (예: 이종현 2017038088): ").split()
    name = student_name[0]
    scores = list(map(int, input(f"{name} 학생의 영어, C-언어, 파이썬 점수를 입력하시오 (예: 80 90 70): ").split()))
    print(f"{name} 학생 영어 점수:{scores[0]}, c-언어 점수:{scores[1]}, 파이썬 점수:{scores[2]}")
    student_info.append([name, student_name[1], scores, {'grade': '', 'avg': -1, 'sum': -1, 'rank':-1}])
  return student_info

def calculate_sum_average(info):
  for student in info:
    total_score = sum(student[2])
    average_score = int(total_score / len(student[2]))
    student[3]['sum'] = total_score
    student[3]['avg'] = average_score


def calculate_grade(info):
  for student in info:
    if student[3]['avg'] == -1 :
      print("평균을 먼저 구하세요")
      return
    if student[3]['avg'] >= 90:
        student[3]['grade'] = "A"
    elif student[3]['avg'] >= 80:
        student[3]['grade'] = "B"
    elif student[3]['avg'] >= 70:
        student[3]['grade'] = "C"
    elif student[3]['avg'] >=60:
        student[3]['grade'] = "D"
    else:
        student[3]['grade'] = "F"


def calculate_rank(info):
  if info[0][3]['avg'] == -1 :
    print("평균을 먼저 구하세요")
    return

  info.sort(key=lambda x: x[3]['avg'], reverse=True)
  for index, student in enumerate(info):
    student[3]['rank'] = index

def print_results(info):
  check = info[0][3]
  if check['sum'] == -1 or check['avg'] == -1 or check['rank'] == -1 or check['grade'] == '':
    print("결과를 출력 할 수 없습니다")
    return

  for index, item in enumerate(info):
    print('')
    print(f"{item[0]} 학생 총점={item[3]['sum']}, 평균={item[3]['avg']:.2f}, 학점={item[3]['grade']}, 등수={item[3]['rank'] + 1}")
```

```python
student = input_student_info(5)
calculate_sum_average(student)
calculate_grade(student)
calculate_rank(student)
print_results(student)
```
